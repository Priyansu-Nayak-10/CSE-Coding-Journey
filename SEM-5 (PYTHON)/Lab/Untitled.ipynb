```python
from random import randrange as rr
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
n_rolls=6000
rolls=[rr(1,7) for _ in range(n_rolls)]
values,frequencies=np.unique(rolls,return_counts=True)
title=f"Rolling a six side die {n_rolls} times"
sns.set_style('darkgrid')
ax = sns.barplot(x=values, y=frequencies, palette='bright')
for bar, frequency in zip(ax.patches, frequencies):
    x_t=bar.get_x()+bar.get_width()/2
    y_t=bar.get_height()
    t=f"{frequency}\n{frequency/n_rolls:.2%}"
    ax.text(x_t,y_t,t, ha='center', va='bottom')
plt.show()
```
